```python
import matplotlib.pyplot as plt
from statistics import mean 
import numpy as np
import pandas as pd
import math
```

```python
%matplotlib inline 
```

```python
usecols = ["CPU", "LPM", "Transmit", 'Listen', 'Idle_transmit', 'Idle_listen']
# read data from csv
df = pd.read_csv('power3.csv')
# set index as Name
df.set_index("ID", inplace=True)
# Returns a new DataFrame sorted by label
df.info()
df.head(30)
```

```python
colors = ['darkorange', 'crimson', 'cornflowerblue', 'limegreen','turquoise','tomato']
labels = ['CPU', 'LPM', 'Transmit', 'Listen', 'Idle_transmit', 'Idle_listen']

time = df.Time
cpu = df.CPU
lpm = df.LPM
transmit = df.Transmit
listen = df.Listen
i_transmit = df.Idle_transmit
i_listen = df.Idle_listen
```

```python
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
ax = plt.gca()
for row in df.itertuples():
    plt.scatter(x=getattr(row, 'Time'), y=getattr(row, labels[0]), c=colors[0])
    
#     plt.scatter(x=getattr(row, 'Time'), y=getattr(row, 'LPM'), c=colors[1])
#     plt.scatter(x=getattr(row, 'Time'), y=getattr(row, 'Listen'), c=colors[2])
#     plt.scatter(x=getattr(row, 'Time'), y=getattr(row, 'Transmit'), c=colors[3])
plt.plot(time, cpu, c=colors[0], alpha=0.5, label=labels[0])
ax.xaxis.grid(True, linestyle='--')
ax.yaxis.grid(True, linestyle='--')
plt.xlabel("Time", fontsize=14)
plt.ylabel(labels[0], fontsize=14)
plt.title( 'CPU consumption', fontsize=24, y=1.02)
```
