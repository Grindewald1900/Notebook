```python
import matplotlib.pyplot as plt
from statistics import mean 
import numpy as np 
import pandas as pd
import math
```

```python
usecols = ["ID", "Size", "Temperature"]
# read data from csv
df = pd.read_csv('data.csv')
# Returns a new DataFrame sorted by label
df.info()
df.head(10)
```

```python
labels = ["ID", "Size", "Temperature"]
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
ax = plt.gca()
for row in df.itertuples():
    plt.scatter(x=getattr(row, 'Size'), y=getattr(row, 'Temperature'), c='turquoise')
    
plt.plot(df.Size, df.Temperature, c='turquoise', alpha=0.5, label='Temperature')
ax.xaxis.grid(True, linestyle='--')
ax.yaxis.grid(True, linestyle='--')
plt.xlabel("Size", fontsize=14)
plt.ylabel('Temperature', fontsize=14)
plt.title('Temperature evolution with array size increasing', fontsize=24, y=1.02)
```
